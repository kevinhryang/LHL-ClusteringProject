```python
#age frequency
plt.figure(figsize=(6,4))
plt.title("Ages Frequency")
sns.axes_style("dark")
sns.violinplot(y=table["age"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(3,5,1)
sns.boxplot(y=table["avg(amount/tran)"], color="red")
plt.subplot(3,5,3)
sns.boxplot(y=table["income"])
plt.subplot(3,5,5)
sns.boxplot(y=table["credit_balance"])
plt.subplot(3,5,6)
sns.boxplot(y=table["savings_balance"])
plt.subplot(3,5,8)
sns.boxplot(y=table["checking_balance"])
plt.subplot(3,5,10)
sns.boxplot(y=table["age"])
plt.subplot(3,5,12)
sns.boxplot(y=table["years_with_bank"])
plt.subplot(3,5,14)
sns.boxplot(y=table["nbr_children"])

plt.show()
```

```python
gender = table.gender.value_counts(normalize=True) * 100
sns.set_style("darkgrid")
plt.figure(figsize=(6,2))
sns.barplot(x=gender.index, y=gender.values)
plt.show()
```

```python
age_under18 = table.age[table.age <= 18]
age18_29 = table.age[(table.age <= 29) & (table.age >= 18)]
age30_39 = table.age[(table.age <= 39) & (table.age >= 30)]
age40_49 = table.age[(table.age <= 49) & (table.age >= 40)]
age50_59 = table.age[(table.age <= 59) & (table.age >= 50)]
age_above60 = table.age[table.age >= 60]

x = ["under 18","18-29","30-39","40-49","50-59","above 60"]
y = [len(age_under18)/len(table['age']) * 100,len(age18_29)/len(table['age'])* 100,len(age30_39)/len(table['age'])* 100,len(age40_49)/len(table['age'])*100,len(age50_59)/len(table['age'])*100,len(age_above60)/len(table['age'])*100]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("Percentage of Customers per age range")
plt.xlabel("Age ranges")
plt.ylabel("Percent of Customers")
plt.show()
```

```python
avg_100 = table["avg(amount/tran)"][table["avg(amount/tran)"] <= 100]
avg_200 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 100) & (table["avg(amount/tran)"] <= 200)]
avg_300 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 200) & (table["avg(amount/tran)"] <= 300)]
avg_400 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 300) & (table["avg(amount/tran)"] <= 400)]
avg_500 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 400) & (table["avg(amount/tran)"] <= 600)]
avg_600 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 600) & (table["avg(amount/tran)"] <= 700)]
avg_700 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 700) & (table["avg(amount/tran)"] <= 4050)]

#avg_4050 = table["avg(amount/tran)"][(table["avg(amount/tran)"] > 3500) & (table["avg(amount/tran)"] <= 4050)]

avg_x = ["under 100", "100 - 200", "200 - 300", "300 - 400", "400 - 500", "500 - 600", "600 - 700", "over 700"]
avg_y = [len(avg_100)/len(table['avg(amount/tran)'])*100, len(avg_200)/len(table['avg(amount/tran)'])*100, len(avg_300)/len(table['avg(amount/tran)'])*100, len(avg_400)/len(table['avg(amount/tran)'])*100, len(avg_500)/len(table['avg(amount/tran)'])*100, len(avg_600)/len(table['avg(amount/tran)'])*100, len(avg_700)/len(table['avg(amount/tran)'])*100]

plt.figure(figsize=(15,6))
sns.barplot(x=avg_x, y=avg_y, palette="nipy_spectral_r")
plt.title("Percentage of Customers ranging by average amount per transactions")
plt.xlabel("Average amount per transactions ranges")
plt.ylabel("Percentage of Customers")
plt.show()
```
